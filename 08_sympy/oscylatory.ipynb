```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
from IPython.display import display
sp.init_printing()
```

```python
x = sp.Function('x')
t = sp.Symbol('t')
w0 = sp.Symbol('omega_0', positive=True, real=True)
v0 = sp.Symbol('v_0', real=True)
x0 = sp.Symbol('x_0', real=True)

tau = sp.Symbol('tau', real=True)
beta = sp.Integer(1)/(2*tau)

x(t).diff(t, t)
```

```python
# Wprowadzam tau jako odwrotnosc tlumienia, bo lepiej podstawić zero zamiast inf
osc_tl = sp.Eq(x(t).diff(t, t) + tau * x(t).diff(t) + w0**2 * x(t), 0)

osc_tl.subs(tau, 0)
```

```python
sol_tl = sp.dsolve(osc_tl, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
display(sol_tl)
sp.plot(sol_tl.rhs.subs({x0: 0, w0: 1, v0: 1, tau: 0.1}), (t, 0, 40))
```

```python
sol = sp.dsolve(osc_tl.subs(tau, 0), x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
display(sol)
sp.plot(sol.rhs.subs({x0: 0, w0: 1, v0: 1}), (t, 0, 40))
```
